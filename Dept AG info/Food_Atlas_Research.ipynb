```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# orig_working_directory = os.getcwd()
# print(orig_working_directory)
file_path = 'Food_Atlas_Reserch.csv'

food_desert_2010 = pd.read_csv(file_path)
```

```python
food_desert_2010.columns
```

```python
# Original file has 65 columns. I pair this down to remove columns for 10 and 20 mile impact (Our focus is on rural areas(cities))

food_1nhalf_pd = food_desert_2010[['CensusTract','State','County','LILATracts_1And10','Urban', 'Rural','POP2010', 'UATYP10',
                                   'LILATracts_Vehicle','LATracts_half', 'LATracts1','HUNVFlag','OHU2010','NUMGQTRS', 'PCTGQTRS', 'LowIncomeTracts',
                                    'lapophalf', 'lapophalfshare', 'lalowihalf', 'lalowihalfshare', 'laseniorshalf', 'laseniorshalfshare',
                                    'lapop1', 'lapop1share', 'lalowi1','lalowi1share', 'lakids1', 'lakids1share', 'laseniors1',
                                    'laseniors1share', 'lahunv1', 'lahunv1share']]

food_1nhalf_pd.head(20)
```

```python
US_pop = food_1nhalf_pd['POP2010'].sum()
US_pop
```

```python
DC_data = food_1nhalf_pd.loc[(food_1nhalf_pd['State']=='DC')]
DC_pop = DC_data['POP2010'].sum()
DC_pop
```

```python
State_Pop = food_1nhalf_pd.groupby(food_1nhalf_pd['State'])['POP2010'].sum()
#State_Pop = State_Pop['CensusTract','Pop2010','LILATracts_1And10']
State_Pop = pd.DataFrame(State_Pop)

Mile1n10 = food_1nhalf_pd.loc[(food_1nhalf_pd['Urban'] > 0) &
                             (food_1nhalf_pd['LATracts1'] > 0)]

Mile1n10_pd = Mile1n10.groupby(['State'])


#Mile1n10.set_index('State',inplace=True)

#Mile1n10[['lakids1','laseniors1']]

State_Pop['Urban FD 1 mile']= Mile1n10_pd['POP2010'].sum()
State_Pop['Urban FD % Pop'] = (State_Pop['Urban FD 1 mile']/ State_Pop['POP2010'])*100
State_Pop['Kid Population'] = Mile1n10_pd['lakids1'].sum()
State_Pop['% kids Urban FD'] = State_Pop['Kid Population']/State_Pop['POP2010']*100
State_Pop['Sr Population'] = Mile1n10_pd['laseniors1'].sum()
State_Pop['% Srs Urban FD'] = State_Pop['Sr Population']/State_Pop['POP2010']*100


#(Mile1n10['lakids1']/ State_Pop['POP2010'])
# State_Pop['Elderly Population'] = (Mile1n10['laseniors1']/ State_Pop['POP2010'])


State_Pop
```

## Sort and pull out 5 States to for comparison to DC.

```python
STATES = State_Pop.reset_index()
STATES
```

```python
Metro_pop = STATES.loc[(STATES['State'] == 'DC')|
                         (STATES['State'] == 'MD')|
                         (STATES['State'] == 'VA')| 
                         (STATES['State'] == 'NY')|                         
                         (STATES['State'] == 'PA')]

Metro_pop
```

```python
Metro_pop = Metro_pop[['State','POP2010','Urban FD % Pop','% kids Urban FD','% Srs Urban FD']]
Metro_pop
```

```python
Metro_pop["Urban FD % Pop"] = Metro_pop["Urban FD % Pop"].map("{0:,.2f}%".format)
Metro_pop["% kids Urban FD"] = Metro_pop["% kids Urban FD"].map("{0:,.2f}%".format)
Metro_pop["% Srs Urban FD"] = Metro_pop["% Srs Urban FD"].map("{0:,.2f}%".format)

Metro_pop
```

```python
Metro_pop.rename(columns = {'POP2010':'2010 Pop.', 'Urban FD % Pop':'Urban Food Desert % Pop', 
                              '% kids Urban FD':'% Kids in Urban FD','% Srs Urban FD':'% Srs in Urban Food Desert' }, inplace = True)
Metro_pop
```
