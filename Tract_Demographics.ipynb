```python
demo_census_2014_2018 = converted_deserts_census_all_years_df[['Total Population','Total Black or African American alone',
                                                              'Total American Indian and Alaska Native alone',
                                                               'Total White alone',
                                                              'Total Asian alone',
                                                              'Total Native Hawaiian and Other Pacific Island alone',
                                                             'Some other race alone','Median Age','Gini Index of Inequality', 
                                                           'county','tract', 'year']]

demo_census_tract_11100 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']== 11100,:]
demo_census_tract_9601 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==9601,:]
demo_census_tract_7806 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==7806,:]
demo_census_tract_7809 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']== 7809,:]
demo_census_tract_9501 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==9501,:]
demo_census_tract_2102 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==2102,:]
demo_census_tract_7301 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==7301,:]
demo_census_tract_7401 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==7401,:]
demo_census_tract_7603 = demo_census_2014_2018.loc[demo_census_2014_2018['tract']==7603,:]


```

### Tract 11100 Race Demographics
This tract is no longer a food desert 

```python
#Manipulating Data to Plot
year_x_axsis= demo_census_tract_11100['year']
black_11100 = demo_census_tract_11100['Total Black or African American alone']
native_11100 = demo_census_tract_11100['Total American Indian and Alaska Native alone']
white_11100 = demo_census_tract_11100['Total White alone']
asian_11100 = demo_census_tract_11100['Total Asian alone']
other_11100 = demo_census_tract_11100['Some other race alone']


totals = [a+b+c+d+e for a,b,c,d,e in zip(demo_census_tract_11100['Total Black or African American alone'],
                                         demo_census_tract_11100['Total American Indian and Alaska Native alone'], 
                                         demo_census_tract_11100['Total White alone'],
                                        demo_census_tract_11100['Total Asian alone'],
                                        demo_census_tract_11100['Some other race alone'])]
greenBars = [i / j * 100 for i,j in zip(demo_census_tract_11100['Total Black or African American alone'], totals)]           
orangeBars = [i / j * 100 for i,j in zip(demo_census_tract_11100['Total American Indian and Alaska Native alone'], totals)]
blueBars = [i / j * 100 for i,j in zip(demo_census_tract_11100['Total White alone'], totals)]
yellowBars = [i / j * 100 for i,j in zip(demo_census_tract_11100['Total Asian alone'], totals)]
redBars =[i / j * 100 for i,j in zip(demo_census_tract_11100['Some other race alone'], totals)] 
barWidth = 1

#Create Bars
plt.figure(figsize=(10,10))
plt.bar(year_x_axsis, greenBars, label= 'Total Black or African American alone', color='#b5ffb9', edgecolor='white')
plt.bar(year_x_axsis, orangeBars,label= 'Total American Indian and Alaska Native alone', bottom=greenBars, color='#f9bc86', edgecolor='white')
plt.bar(year_x_axsis, blueBars,label= 'Total White alone', bottom=[i+j for i,j in zip(greenBars, orangeBars)], color='#a3acff', edgecolor='white')
plt.bar(year_x_axsis, yellowBars,label= 'Total Asian alone', bottom=[i+j+k for i,j,k in zip(greenBars, orangeBars,blueBars)], color='#FFFF99', edgecolor='white')
plt.bar(year_x_axsis, redBars,label= 'Some other race alone', bottom=[i+j+k+l for i,j,k,l in zip(greenBars, orangeBars,blueBars,yellowBars)], color='#FA8072', edgecolor='white')

#Design Edits 
plt.xticks(year_x_axsis)
plt.xlabel("Year")
plt.ylabel("Percent of Population")
plt.title("Tract 11100 Race Demographics ")
leg = plt.legend();

# Save graphic
plt.savefig('Graphs/Tract_011100_Race_StackedBar.png')
```
